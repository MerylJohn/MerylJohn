```python
import pandas as pd
dataframe=pd.read_csv("titanic.csv")
```

```python
dataframe.head()
```

```python
x=dataframe.drop('Survived',1)
y=dataframe.Survived
```

```python
print(x)
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=50)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree=DecisionTreeClassifier()
tree=tree.fit(X_train,y_train)
predictedy=tree.predict(X_test)
```

```python
from sklearn import metrics
accuracy=metrics.accuracy_score(y_test,predictedy)
print(accuracy)
```

```python
from sklearn.tree import export_graphviz
from io import StringIO
```

```python
dodata=StringIO()
export_graphviz(tree,out_file=dodata,filled=True, rounded=True, special_characters=True,class_names=['0','1'])
print(dodata.getvalue())
```

```python
import pydotplus
graph=pydotplus.graph_from_dot_data(dodata.getvalue())
graph.write_png("tree for titanic png")
```
